## Geospatial analysis

```python
# Load the geoparquet file
gdf = gpd.read_parquet('./data/nz-building-outlines.parquet')
gdf.head()
```

```python
# filter the data based on attribute
gdf_filtered = gdf[gdf['town_city'] == 'Marton']

# buffer geometry by 10 meters
gdf_filtered['geometry'] = gdf_filtered.buffer(10)

# load only the first 10000 rows
gdf_filtered = gdf_filtered.head(10000)
```

## visualization

```python
# plot the data
gdf_filtered.plot()
```

```python
# save the data
gdf_filtered.to_file('./data/output/marton-buildings-buffered.geojson', driver='GeoJSON')
```
